## Create a cursor to connect to the database

```python
# Create readonly database connection and cursor
import sqlite3
conn = sqlite3.connect('file:../data/BTCUSDT.db?mode=ro', uri=True)
curs = conn.cursor()
```

## Do some database stuff

```python
# Get all tables listed in the sqlite_master database
curs.execute("SELECT name FROM sqlite_master WHERE type = 'table'").fetchall()
```

```python
# Access the tables one by one
for row in curs.execute("SELECT name FROM sqlite_master WHERE type = 'table'"):
    print(row)
```

```python
# Show the columns in a particular table
for row in curs.execute("PRAGMA table_info(_1d)"):
    print(row)
```

```python
# Connect to database and create a dataframe from last 10 rows (Descending)
latest = sqlite3.connect('file:../data/BTCUSDT.db?mode=ro', uri=True)
df_latest=pd.read_sql_query("SELECT * FROM _1d ORDER BY openTime DESC LIMIT 10", latest)
conn.close()

# After the dataframe has been build, reorder the dates so that the last date is the last.
# First change the Unix time to human readable time
df_latest['openTime']=(pd.to_datetime(df_latest['openTime'],unit='s'))

# This line sorts the openTime ascending (but not necessary anymore with the next line)
# df_latest.sort_values(by=['openTime'], ascending=True)

# This line reorders the dataframe and rebuilds (replaces) the old dataframe (variable)
df_latest = df_latest.set_index(['openTime']).sort_values(by=['openTime'], ascending=True)

df_latest
```
